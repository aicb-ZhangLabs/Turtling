```python
import pygraphviz as PG
import numpy as np
from IPython.display import Image

colors = ["#800000", "#A52A2A", "#DC143C", "#FF6347", "#CD5C5C",\
          "#E9967A", "#FFA07A", "#FF8C00", "#FFD700", "#DAA520",\
          "#BDB76B", "#808000", "#9ACD32", "#20B2AA", "#4682B4",\
          "#8A2BE2", "#9400D3", "#FFB6C1", "#F5DEB3", "#DEB887"]

level1 = list(range(20))
level2 = np.random.randint(20, size=50)

G = PG.AGraph(directed=True)

G.add_node("0", label="", shape="circle", style="bold", color="#000000")

for i in range(len(level1)):
     G.add_node("1_"+str(i), label="", shape="circle", style="bold", color=colors[i])
for i in range(len(level1)):
     G.add_edge('0',"1_"+str(level1[i]))

for i in range(len(level2)):
     G.add_node("2_"+str(i), label="", shape="square", style="bold", color=colors[level2[i]])

for i in range(len(level2)):
    G.add_edge("1_"+str(level2[i]), "2_"+str(i))

G.layout(prog='dot')
Image(G.draw(format='png'))
```
